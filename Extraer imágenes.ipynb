```python

#### importamos librerías

import requests
import pandas as pd

API_KEY =  "PONER EL API KEY"
```

```python
#ESTA ES UNA FUNCIÓN PARA DESCARGAR UNA IMAGEN POR CADA UNO DE LOS PUNTOS QUE TENEMOS EN LA BASE DE DATOS GENERADA EN R
# ES NECESARIO CREAR UNA CARPETA QUE SE LLAME IMAGENES EN EL QUE 

def descargar_imagen(lat, lon, nombre):
    url = f"https://maps.googleapis.com/maps/api/streetview?size=640x640&location={lat},{lon}&key={API_KEY}"
    r = requests.get(url)
    with open(f"imagenes/{nombre}.jpg", "wb") as f:
        f.write(r.content)
```

```python
### LEEMOS LAS BASE DE DATOS QUE GENERAMOS EN R

df = pd.read_csv("puntos_cuauhtemoc.csv")
```

```python

# DESCARGAR UNA IMAGEN POR CADA PUNTO

for i, row in df.iterrows():
    descargar_imagen(row["lat"], row["lon"], i)
```
